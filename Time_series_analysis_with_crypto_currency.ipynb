```python
def create_dataset(dataset, column_name, time_steps=1):
    """
    Create a time series dataset for the specified column.

    Args:
    dataset (pd.DataFrame): Input dataset (normalized or raw).
    column_name (str): The name of the column to use for creating the dataset.
    time_steps (int): The number of time steps to consider for each input sample.

    Returns:
    np.array: X (features), time series input data.
    np.array: y (labels), time series target data.
    """
    X, y = [], []

    # Extract the specified column as a NumPy array
    column_data = dataset[column_name].values

    # Generate sequences
    for i in range(len(column_data) - time_steps):
        X.append(column_data[i:(i + time_steps)])
        y.append(column_data[i + time_steps])

    return np.array(X), np.array(y)
```

# **Isolation Forest (Machine Learning Method)**
An advanced machine learning algorithm that identifies outliers by isolating them in a random forest model.

```python
from sklearn.ensemble import IsolationForest

# Train Isolation Forest for anomaly detection
model = IsolationForest(contamination=0.05)
df['Anomaly (IF)'] = model.fit_predict(df[['Close(BTC-USD)']])

# Flag anomalies (1 for normal, -1 for anomaly)
df['Anomaly (IF)'] = np.where(df['Anomaly (IF)'] == -1, 1, 0)

# Preview flagged anomalies
print(df[['Date', 'Close(BTC-USD)', 'Anomaly (IF)']].head(10))
```

# **Z-Score for Anomaly Detection**
This method will help you flag data points that deviate significantly from the mean, indicating potential outliers or anomalies.

```python
import numpy as np

# Calculate Z-score for detecting anomalies
df['Z-Score'] = (df['Close(BTC-USD)'] - df['Close(BTC-USD)'].mean()) / df['Close(BTC-USD)'].std()

# Flag anomalies where Z-score is greater than a threshold (e.g., 3)
threshold = 3
df['Anomaly'] = np.where(df['Z-Score'].abs() > threshold, 1, 0)

# Preview flagged anomalies
print(df[['Date', 'Close(BTC-USD)', 'Z-Score', 'Anomaly']].head(10))
```

# **IQR (Interquartile Range) Method**
The IQR method is another way to flag anomalies based on the spread of the data.

```python
# Calculate IQR for anomaly detection
Q1 = df['Close(BTC-USD)'].quantile(0.25)
Q3 = df['Close(BTC-USD)'].quantile(0.75)
IQR = Q3 - Q1

# Define anomaly detection thresholds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Flag anomalies based on IQR
df['IQR Anomaly'] = np.where((df['Close(BTC-USD)'] < lower_bound) | (df['Close(BTC-USD)'] > upper_bound), 1, 0)

# Preview flagged anomalies
print(df[['Date', 'Close(BTC-USD)', 'IQR Anomaly']].head(10))
```

```python
import re
```

```python
def cleanTwt(twt):
    twt = re.sub("#bitcoin", 'bitcoin', twt) # removes the '#' from bitcoin
    twt = re.sub("#Bitcoin", 'Bitcoin', twt) # removes the '#' from Bitcoin
    twt = re.sub('#[A-Za-z0-9]+', '', twt) # removes any string with a '#'
    twt = re.sub('\\n', '', twt) # removes the '\n' string
    twt = re.sub('https:\/\/\S+', '', twt) # removes any hyperlinks
    return twt
```

```python
df['cleaned_tweets'] = df['tweets'].apply(cleanTwt)
df
```

```python
from textblob import TextBlob
```

```python
def getSubjectivity(twt):
    return TextBlob(twt).sentiment.subjectivity

# create a function to get the polarity
def getPolarity(twt):
    return TextBlob(twt).sentiment.polarity

# create two new columns called "Subjectivity" & "Polarity"
df['subjectivity'] = df['cleaned_tweets'].apply(getSubjectivity)
df['polarity'] = df['cleaned_tweets'].apply(getPolarity)
```

```python
def getSentiment(score):
    if score < 0:
        return "negative"
    elif score == 0:
        return "neutral"
    else:
        return "positive"
```

```python
df['sentiment'] = df['polarity'].apply(getSentiment)

# show the data
df
```
