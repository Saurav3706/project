```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
plt.style.use("fivethirtyeight")
%matplotlib inline

# Import yfinance and datetime for data retrieval
import yfinance as yf
from datetime import datetime

# Define the cryptocurrency symbol for Bitcoin
crypto_list = ['BTC-USD']  # Bitcoin to USD

# Set the start and end dates
end = datetime.now()
start = datetime(2010, 7, 17)  # Bitcoin's earliest price data on Yahoo Finance

# Create an empty list to store the data
crypto_data_list = []

# Loop through the cryptocurrency list to fetch data
for crypto in crypto_list:
    # Download data using yfinance
    data = yf.download(crypto, start=start, end=end)
    data["crypto_name"] = crypto  # Add a column for the cryptocurrency name
    crypto_data_list.append(data)

# Concatenate the DataFrames into a single DataFrame
df = pd.concat(crypto_data_list, axis=0)

# Display the last 10 rows of the DataFrame
print(df.tail(10))

# Save the data to a CSV file for future use (optional)
df.to_csv("bitcoin_data.csv", index=True)

# Plot the historical closing price of Bitcoin
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x=df.index, y='Close', label='Bitcoin (BTC-USD)')
plt.title("Bitcoin Historical Closing Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()

```

```python
def create_dataset(dataset, column_name, time_steps=1):
    """
    Create a time series dataset for the specified column.

    Args:
    dataset (pd.DataFrame): Input dataset (normalized or raw).
    column_name (str): The name of the column to use for creating the dataset.
    time_steps (int): The number of time steps to consider for each input sample.

    Returns:
    np.array: X (features), time series input data.
    np.array: y (labels), time series target data.
    """
    X, y = [], []

    # Extract the specified column as a NumPy array
    column_data = dataset[column_name].values

    # Generate sequences
    for i in range(len(column_data) - time_steps):
        X.append(column_data[i:(i + time_steps)])
        y.append(column_data[i + time_steps])

    return np.array(X), np.array(y)
```

```python
# Updated create_dataset function to accept NumPy arrays directly
def create_dataset(dataset, time_steps=1):
    X, y = [], []
    for i in range(len(dataset) - time_steps):
        a = dataset[i:(i + time_steps)]  # Keep it as a NumPy array slice
        X.append(a)
        y.append(dataset[i + time_steps])  # Use dataset directly without .values
    return np.array(X), np.array(y)
```

```python
# Set the number of time steps
time_steps = 40  # You can adjust this based on your needs

# Create dataset for Bitcoin prices (already a NumPy array)
X_bitcoin, y_bitcoin = create_dataset(bitcoin_prices, time_steps)

# Split the dataset into training and testing sets
split_ratio = 0.8  # 80% for training, 20% for testing
split_bitcoin = int(split_ratio * len(X_bitcoin))

# Training and testing sets for Bitcoin
X_train_bitcoin, y_train_bitcoin = X_bitcoin[:split_bitcoin], y_bitcoin[:split_bitcoin]
X_test_bitcoin, y_test_bitcoin = X_bitcoin[split_bitcoin:], y_bitcoin[split_bitcoin:]

# Print dataset shapes to confirm
print(f"X_train_bitcoin shape: {X_train_bitcoin.shape}")
print(f"y_train_bitcoin shape: {y_train_bitcoin.shape}")
print(f"X_test_bitcoin shape: {X_test_bitcoin.shape}")
print(f"y_test_bitcoin shape: {y_test_bitcoin.shape}")
```

```python
# Reshape the datasets to (samples, time steps * features)
X_train_bitcoin = X_train_bitcoin.reshape(X_train_bitcoin.shape[0], -1)
X_test_bitcoin = X_test_bitcoin.reshape(X_test_bitcoin.shape[0], -1)

# Print the reshaped dataset shapes to confirm
print(f"X_train_bitcoin reshaped: {X_train_bitcoin.shape}")
print(f"X_test_bitcoin reshaped: {X_test_bitcoin.shape}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

# Linear Regression model
model = LinearRegression()

# Convert y_train_bitcoin to a 1D array (to avoid the DataConversionWarning)
y_train_bitcoin = y_train_bitcoin.ravel()
y_test_bitcoin = y_test_bitcoin.ravel()

# Train the Linear Regression model
model.fit(X_train_bitcoin, y_train_bitcoin)

# Make predictions for Bitcoin
predictions_bitcoin = model.predict(X_test_bitcoin)

# Evaluate performance for Bitcoin
mse_bitcoin = mean_squared_error(y_test_bitcoin, predictions_bitcoin)

# Print the results
print("Model: Linear Regression")
print("Mean Squared Error for Bitcoin:", mse_bitcoin)
```
